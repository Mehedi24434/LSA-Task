### Egovernance

```python
# Data
categories = ['E-Government Index', 'E-Participation Index', 'Online Service Index', 'Human Capital Index', 'Telecommunication Infrastructure Index']
countries = ['Bangladesh eGovt rank 100', 'Sri Lanka eGovt rank 98', 'Cambodia eGovt rank 120']
values = [
    [0.65696, 0.6164, 0.73744, 0.58337, 0.6501],  # Bangladesh
    [0.66668, 0.411, 0.54937, 0.65704, 0.79359],  # Sri Lanka
    [0.57536, 0.3151, 0.45034, 0.51489, 0.76094]  # Cambodia
]

# Number of variables
num_vars = len(categories)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Complete the loop
values = [values[i] + values[i][:1] for i in range(len(countries))]
angles += angles[:1]

# Create radar chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Add each country's data to the radar chart
for i in range(len(countries)):
    ax.fill(angles, values[i], alpha=0.25, label=countries[i])
    ax.plot(angles, values[i], linewidth=2, linestyle='solid')

    # Annotate the values with slight offsets
    for j in range(num_vars):
        if j > 0:
            offset_angle = angles[j] + 0.1  # Offset angle slightly
            offset_value = values[i][j] + 0.03  # Offset value slightly
            ax.text(offset_angle, offset_value, f'{values[i][j]:.5f}', ha='center', va='center', fontsize=9, color='black')

# Labels and title
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=10)
plt.title('E-Government Performance Comparison as of Year 2024', size=10, color='black', weight='bold')

# Adjust legend position
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))  # Move the legend further right

# Show the plot
plt.show()
```

### Digital Payment

```python
# Data
data = {
    'Country': ['Bangladesh', 'Bangladesh', 'Bangladesh',
                'Sri Lanka', 'Sri Lanka', 'Sri Lanka',
                'Cambodia', 'Cambodia', 'Cambodia'],
    'Code': ['BGD', 'BGD', 'BGD', 'LKA', 'LKA', 'LKA', 'KHM', 'KHM', 'KHM'],
    'Year': [2021, 2017, 2014, 2021, 2017, 2014, 2021, 2017, 2014],
    'Value': [33.56, 21.21, 4.58, 47.51, 44.2, 23.57, 28.62, 16.08, 19.25]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Pivoting the DataFrame to get years as index and countries as columns
pivot_df = df.pivot(index='Year', columns='Country', values='Value')

# Calculating the average
average_value = 53.3548702185792

# Plotting the line chart
plt.figure(figsize=(10, 6))
for country in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[country], marker='o', label=country)
    
    # Adding values on the plot
    for x, y in zip(pivot_df.index, pivot_df[country]):
        plt.text(x, y, f'{y:.2f}', ha='center', va='bottom')  # Display values above points

# Adding a horizontal line for the average
plt.axhline(y=average_value, color='r', linestyle='--', label='Average Value')

# Customizing the plot
plt.title('Made or received a digital payment (% age 15+)')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(pivot_df.index)
plt.legend()
plt.grid(False)  # Remove grid lines
plt.ylim(0, 60)  # Adjust y-axis limit based on data
plt.tight_layout()

# Show the plot
plt.show()
```

### Mobile Cellular Subscriptions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data preparation
data = {
    'Country Name': ['Bangladesh', 'Sri Lanka', 'Cambodia'],
    'Mobile Cellular Subscriptions': [191000000, 30000000, 22100000],
    'Times Total Population': [1.12780011881581, 12.9636559583988, 1.2529196287208],
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Setting the country names as the index
df.set_index('Country Name', inplace=True)

# Bar width
bar_width = 0.35

# X locations for the groups
x = np.arange(len(df.index))

# Creating a figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Creating the bar chart for Mobile Cellular Subscriptions
bars1 = ax1.bar(x, df['Mobile Cellular Subscriptions'], width=bar_width, color='b', alpha=0.7, label='Mobile Cellular Subscriptions')

# Setting labels and title for the first y-axis
ax1.set_xlabel('Country Name')
ax1.set_ylabel('Number of Mobile Cellular Subscriptions', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_title('Number of Mobile Cellular Subscriptions and Times Total Population by Country by Year 2024')

# Adding values on top of the first set of bars
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, height, f'{height:,}', ha='center', va='bottom')

# Creating a second y-axis for Times Total Population
ax2 = ax1.twinx()
bars2 = ax2.bar(x + bar_width, df['Times Total Population'], width=bar_width, color='r', alpha=0.7, label='Times Total Population')

# Setting labels for the second y-axis
ax2.set_ylabel('Times Total Population', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Adding values on top of the second set of bars
for bar in bars2:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}x', ha='center', va='bottom')

# Customizing x-ticks
ax1.set_xticks(x + bar_width / 2)  # Centering the x-ticks
ax1.set_xticklabels(df.index)

# Adding legends
ax1.legend(loc='upper right')
ax2.legend(loc='right')

# Show the plot
plt.tight_layout()
plt.show()
```

### Digital Literacy Rates by Country'

```python
import matplotlib.pyplot as plt

# Data for digital literacy rates
countries = ['Bangladesh', 'Sri Lanka', 'Cambodia']
digital_literacy_rates = [23, 36, 30]
netherlands_finland_rate = 79

# Creating the bar plot for Bangladesh, Sri Lanka, and Cambodia
plt.figure(figsize=(8, 5))
bars = plt.bar(countries, digital_literacy_rates, color=['orange', 'green', 'blue'])

# Adding values on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height}%', ha='center', va='bottom')

# Adding a line for Netherlands and Finland
plt.axhline(y=netherlands_finland_rate, color='purple', linestyle='--', label='Netherlands & Finland (79%)')

# Customizing the plot
plt.title('Digital Literacy Rates by Country', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Digital Literacy Rate (%)', fontsize=12)
plt.ylim(0, 100)  # Set the y-axis limit from 0 to 100
plt.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```
